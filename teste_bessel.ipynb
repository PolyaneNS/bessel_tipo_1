```
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib

#criar lista
lis = []
i = 24 #tamanho dos números
j = 0

while j < i :
      lis.append(j)
      j+=1
print("lista",lis)

#função de bessel de ordem zero 
seno = np.sin(lis); #para o tipo 1 da função de Bessel
i_n = seno/lis;

epsilon_0 = 8.854*10**12; #F/m
k_b = 1.380*(1/10**23); #J/K
t = 1000; #K
n_e = 10**19; #m
e = 1.6*(1/10**19); #C
r = 10**(-15)
lambd = (epsilon_0*k_b*t)/(n_e*e**2);
lambda_d = np.sqrt(lambd);

#função de bessel para as condições de contorno

i_0 = i_n*r*lambda_d
print("i_0",i_0)

#print(i_n)
fig, ax = plt.subplots()
plt.plot(i_0)
ax.set(title="Função esférica de Bessel (Tipo 1) de ordem zero", xlabel="x", ylabel="i_0(x)")
plt.show()


from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
import random


fig = pyplot.figure()
ax = Axes3D(fig)

sequence_containing_x_vals = list(range(0, 24))
sequence_containing_y_vals = i_0
sequence_containing_z_vals = list(range(0, 24))

random.shuffle(sequence_containing_x_vals)
random.shuffle(sequence_containing_y_vals)
random.shuffle(sequence_containing_z_vals)

ax.plot(sequence_containing_x_vals, sequence_containing_z_vals,sequence_containing_y_vals)
ax.set(title="Função esférica de Bessel (Tipo 1) de ordem zero", xlabel="Variavel x", ylabel="i")
pyplot.show()

```
