```
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib
 
#criar lista
lis = []
i = 24 #tamanho dos números
j = 1  # iníício da contagem
 
while j < i :
      lis.append(j)
      j+=1
print("lista criada :",lis)
 
#função de bessel de ordem zero 
seno = np.sin(lis); #para o tipo 1 da função de Bessel
i_n = seno/lis; #uma funçãão seno
 
#valores de constante e suas unidades 
epsilon_0 = 8.854*10**12; #F/m
k_b = 1.380*(1/10**23); #J/K
t = 1000; #K
n_e = 10**19; #m
e = 1.6*(1/10**19); #C
r = 5*10**(-9) #cm ou 59 pm {considerando um  gás Hidrogeio(H)} 
lambd = (epsilon_0*k_b*t)/(n_e*e**2);
lambda_d = np.sqrt(lambd);
 
#função de bessel para as condições de contorno
 
i_0 = i_n*r*lambda_d
print("lista de i_0 :", i_0)
```

```
from mpl_toolkits.mplot3d import Axes3D
import random


#print(i_n)
fig, ax = plt.subplots()
plt.plot(i_0)
ax.set(title="Função esférica de Bessel (Tipo 1) de ordem zero", xlabel="x", ylabel="i_0(x)")
plt.show()

```

A equação de Bessel tem como sua resolução as chamadas funções de Bessel. Há dois tipo de função, primeiro e segundo tipo, a função de bessel de primeiro tipo é quando suas soluções são finitas na origem. Já a função de Bessel de segundo tipo são soluções onde há um singularidade na origem. Acredito que a utilizaçõa da equação de Bessel seja para descrever a distribuição de fluxo em função da distancia radial.
Utilizando a ferramenta Python, esbocei gráficos para descrever  o sistema do plasma. Utilizei parâmetros onde define-se um plasma de laboratório.

```

fig = plt.figure()
ax = Axes3D(fig)




x_v = list(range(1, 24))
y_v= i_0
z_v = list(range(1, 24))


random.shuffle(x_v)
random.shuffle(y_v)
random.shuffle(z_v)

ax.plot(x_v, z_v, y_v)

ax.set(title="Função esférica de Bessel (Tipo 1) de ordem zero", xlabel="Variavel x", ylabel="i")
plt.show()

```
